#### Utilizando dicionários
Exemplos de como utilizar algumas das estruturas que vimos para facilitar a organização dos dados.

```python
imgs_dict = {}
labels = [0, 1, 2, 0, 2]
paths = ['img1.png', 'img2.png', 'img3.png', 'img4.png', 'img5.png']
classes = ['cachorro', 'gato', 'cachorro', 'gato', 'carro']

for idx, l in enumerate(labels):
    
    if l not in imgs_dict.keys():
        imgs_dict[l] = []
        
    imgs_dict[l].append([paths[idx], classes[idx]])
```

```python
imgs_dict
```

```python
imgs_dict[0][0]
```
